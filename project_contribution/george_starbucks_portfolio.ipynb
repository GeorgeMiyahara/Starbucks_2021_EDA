# Ownership Type Analysis
Examine the distribution of stores by 'ownershipTypeCode' to understand the mix of company-owned and franchised stores.


```python
num_ownership_type = df['ownershipTypeCode' ].unique()
num_ownership_type
```

- Corporations (CO): There are 14,141 instances of businesses with a "Corporation" ownership type. This suggests that corporations are a common ownership structure within your dataset. Corporations are typically larger entities and may have more complex organizational structures.
- Limited Liability Companies (LS): There are 12,106 instances of businesses with a "Limited Liability Company" ownership type. LLCs are a popular choice for smaller businesses due to their flexibility and limited liability protection. The high count of LLCs suggests a significant presence of small to medium-sized businesses in your dataset.
- Joint Ventures (JV): There are 1,712 instances of businesses with a "Joint Venture" ownership type. Joint ventures are relatively less common than corporations and LLCs. They often involve collaboration between multiple parties for specific projects or ventures.
- Franchises (FR): There are 330 instances of businesses with a "Franchise" ownership type. Franchises are a specialized form of business ownership where individuals or entities purchase the right to operate a business under an established brand. The lower count of franchises suggests that they are less prevalent in your dataset compared to other ownership types.

```python
# Count of each ownership type
ownership_counts = df['ownershipTypeCode'].value_counts()
print(ownership_counts)
```

```python
# Assuming you have a DataFrame 'df' with the ownership type counts
ownership_type_counts = df['ownershipTypeCode'].value_counts()

# Calculate the total count
total_count = ownership_type_counts.sum()

# Calculate the percentage for each ownership type
ownership_type_percentage = (ownership_type_counts / total_count) * 100

# Display the percentages
print(ownership_type_percentage)
```

```python
import matplotlib.pyplot as plt

# Bar chart for ownership type distribution
plt.figure(figsize=(8, 6))
ownership_counts.plot(kind='bar')
plt.xlabel('Ownership Type')
plt.ylabel('Count')
plt.title('Ownership Type Distribution')
plt.show()
```

1. **Company-Owned (CO)**: Approximately 49.99% of Starbucks locations are company-owned. This indicates that Starbucks has a significant number of stores that it operates directly, which is common for large chains.

2. **Licensed Stores (LS)**: About 42.79% of Starbucks locations are licensed stores. These are often located within other businesses (like grocery stores or airports) and operated by a third-party under the Starbucks brand. The high percentage suggests that Starbucks has a substantial presence in such partnerships.

3. **Joint Ventures (JV)**: Joint ventures represent a smaller portion, approximately 6.05%, of Starbucks locations. Joint ventures typically involve Starbucks partnering with another entity to operate stores. This percentage suggests that while joint ventures are part of Starbucks' business model, they are not as prevalent as company-owned or licensed stores.

4. **Franchises (FR)**: Franchised stores make up around 1.17% of Starbucks locations. This implies that Starbucks has a limited number of franchise agreements in place. Starbucks often exercises tight control over its brand, so franchising is less common compared to other ownership models.

Overall, Starbucks uses a diverse ownership model, with a significant focus on company-owned and licensed stores.

```python
import plotly.express as px

# Assuming you have a DataFrame 'grouped_data' with 'countryCode', 'ownershipTypeCode', and 'count' columns
grouped_data = df.groupby(['countryCode', 'ownershipTypeCode']).size().reset_index(name='count')
# Create an interactive bar chart using Plotly Express
fig = px.bar(grouped_data, x='countryCode', y='count', color='ownershipTypeCode',
             labels={'countryCode': 'Country Code', 'count': 'Count', 'ownershipTypeCode': 'Ownership Type'},
             title='Ownership Types by Country')

# Customize the layout (e.g., add hover information)
fig.update_traces(texttemplate='%{y}', textposition='outside')

# Show the interactive plot
fig.show()
```

```python
# Assuming you have a DataFrame 'df' with city and ownershipTypeCode columns

# Group by city and ownership type, then count occurrences
grouped_data = df.groupby(['city', 'ownershipTypeCode']).size().reset_index(name='count')

# Create an interactive bar chart using Plotly Express
fig = px.bar(grouped_data, x='city', y='count', color='ownershipTypeCode',
             labels={'city': 'City', 'count': 'Count', 'ownershipTypeCode': 'Ownership Type'},
             title='Ownership Types by City')

# Customize the layout (e.g., add hover information)
fig.update_traces(texttemplate='%{y}', textposition='outside')

# Show the interactive plot
fig.show()
```
